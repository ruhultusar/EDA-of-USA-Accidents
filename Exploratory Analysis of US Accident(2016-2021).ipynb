### What time of the day are accidents most frequent in?

```python
# accident occured in every hour 

plt.figure(figsize=(10,5))
sns.histplot(accident_df.Start_Time.dt.hour,bins=24);
```

- A high percentage of accidents occur between 2pm to 7pm (probably people in a hurry to get back to work
- Next highest percentage is between 6am tp 10 am.

### Which days of the weeks have the most accidents? 

```python
# accident occured in every days of week

sns.histplot(accident_df.Start_Time.dt.dayofweek, bins=7);
```

- On Saturdays, accidents seem to occur more frequently than on other days of the week. It may be due to the fact that fewer people travel outside on weekends. This leads us to a new question: 
#### Is the distribution of accidents by hour the same on weekends as on weekdays?

```python
# accident occured in Sundays 

sunday_start_time= accident_df.Start_Time[accident_df.Start_Time.dt.dayofweek ==5]
plt.figure(figsize=(10,5))
sns.histplot(sunday_start_time.dt.hour, bins=24);
```

```python
# accident occured in Mondays

monday_start_time= accident_df.Start_Time[accident_df.Start_Time.dt.dayofweek ==6]
plt.figure(figsize=(10,5))
sns.histplot(monday_start_time.dt.hour, bins=24);
```

The time period when accidents happen most frequently is the same on weekends and weekdays. Therefore, although accidents may occur less frequently on weekends than they do during the week, the peak occurrence time is the same as it is on weekdays. This disproves our hypothesis that a high proportion of accidents happen between 2 and 7 o'clock because people are rushing to get back to work.

### Is there any missing value in the dataset?

```python
# accident occured in 2016 and 2017

accident_2016= accident_df[accident_df.Start_Time.dt.year == 2016] 
accident_2017= accident_df[accident_df.Start_Time.dt.year == 2017] 
fig,ax= plt.subplots(1,2,figsize=(10,5))

sns.histplot(accident_2016.Start_Time.dt.month,bins=12, ax=ax[0])
ax[0].set_title("2016");

sns.histplot(accident_2017.Start_Time.dt.month,bins=12, ax=ax[1])
ax[1].set_title("2017");
```

After drilling down more it is being found that data is missing in 2016. Maybe even 2017.  That migth be a reson why the accident rate is much higher in december. 

### Are there more accidents in warmer or colder areas?

```python
# read column which has parenthesis in it's name

Temp= pd.DataFrame(accident_df, columns=['Temperature(F)'])
```

```python
plt.figure(figsize=(10,5))
sns.histplot(Temp, kde= True, bins= np.arange(-25,125,5));
```

According to the aforementioned histogram, the majority of accidents take place while the outside temperature is between 50 and 100 degrees Fahrenheit. This disproves our hypothesis that more accidents happen in the extremely cold weather.

- Now, it is clear that the accident rate is much higher in december is not becasue of cold weather.

### In which weather most accident occured? why?

```python
#count the string value and ploting weather condition

x= accident_df['Weather_Condition'].value_counts().head(15)
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
plt.plot(x,'o-g');
```
